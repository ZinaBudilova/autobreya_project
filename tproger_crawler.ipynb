```python
import time
import requests
from bs4 import BeautifulSoup
import re
from fake_useragent import UserAgent
ua = UserAgent(verify_ssl=False)
headers = {'User-Agent': ua.random}
session = requests.session()
responce = session.get('https://tproger.ru/articles/', headers=headers)
page = responce.text
soup = BeautifulSoup(page, 'html.parser')
articles = soup.find_all('article')
```

```python
import csv
my_data = []
for article in articles:
    time.sleep(3)
    href = article.find('a').attrs['href']
    responce = session.get(href, headers=headers)
    page = responce.text
    soup = BeautifulSoup(page, 'html.parser')
    ps = soup.find_all('p')
    article_text = ''
    for p in ps:
        text = p.text
        article_text = article_text + ' ' + text
    my_data.append([href, article_text])
    myFile = open('articles.csv', 'w', encoding = 'utf-8')
    with myFile:
        writer = csv.writer(myFile)
        writer.writerows(my_data)
```

```python
with open('articles.csv', encoding = 'utf-8') as File:
    reader = csv.DictReader(File, delimiter = ',')
    for row in reader:
        print(row['href'])
```
